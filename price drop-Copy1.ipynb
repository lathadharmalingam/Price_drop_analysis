```python
import requests
```

```python
import bs4
```

```python
import lxml
res = requests.get('https://www.pricebefore.com/price-drops/?category=laptops&price-drop=week')
```

```python
soup = bs4.BeautifulSoup(res.text,"lxml")
```

```python
soup
```

```python
soup.select('.filter-options')
```

```python
for i in soup.select('.filter-options'):
    print(i.text)
```
